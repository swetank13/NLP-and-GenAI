#### 2.3: AI Agents and Agentic AI
- Systems built using LLMs can be divided into two categories:
  1. Workflows (Include RAG Systems and Tool Augmented Systems)
  2. Agents
- Workflows use LLMs but do not have autonomy and complex reasoning.
- `AI Agents` are autonomous software systems that perceive their environment, make decisions, and act to achieve specific goals with minimal human intervention.
- `Agentic AI` refers to advanced AI systems capable of reasoning, planning, and executing complex, multi-step tasks independently, often coordinating with other agents.
- Agentic AI systems will have one or more AI agents in them performing complex tasks.

#### Steps to Build Gen AI and Agentic Applications
- The first step while building Agentic AI apps is to honestly ask this question: Is using LLM really needed, If you can solve the problem efficiently using traditional coding, statistical ML, deep learning, etc., then do not use LLM.
- The steps to build Gen AI or Agentic AI app are as follows:
  - Evaluate if LLM is needed -> Data collection and prep -> Application architecture -> Model training and fine tuning -> Evaluation -> Optimization and deployment -> Compliance and ethical checks -> Monitoring and feedback

#### MCQ
- Gen AI generates, AI Agents use tools and take autonomous actions, Agentic AI plans and acts
- An app summarizes customer feedback, drafts replies, and auto-escalates issues. What type of AI system is this: A system combining Gen AI and Agentic AI.
- What are Large Language Models (LLMs) primarily trained on: Massive corpora of text from books, websites, and documents.
- What makes Agentic AI suitable for business workflows: Its ability to reason, plan, and execute tasks across tools.
